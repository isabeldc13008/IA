```python
# Algoritmo de aprendizaje supervisado (Regresión Lineal)

'''
    La regresión lineal se utiliza en problemas de aprendizaje supervisado donde el objetivo es predecir un valor numérico continuo a partir de una serie de características. 
    Generalmente, se aplica en situaciones en las que se sospecha que existe una relación lineal entre las variables independientes (características) y la variable dependiente (valor objetivo)
'''

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_vals = [6.290303913, 6.278329421, 6.26393719, 6.250503559, 6.247197047, 6.23061541, 6.212712849, 6.315319592, 6.330537796, 6.337816543, 6.252978384, 6.253342459, 6.258716136, 6.256649006]
Y_vals = [-75.56469784, -75.56942353, -75.56334835, -75.56822247, -75.56970975, -75.57554709, -75.57801847, -75.55528241, -75.55338124, -75.54431178, -75.582943, -75.58827777, -75.59773582, -75.61341926]

X = np.array(X_vals).reshape(-1, 1)  # Convertir X_vals en un array 2D
y = np.array(Y_vals)  # Convertir Y_vals en un array de NumPy

# Dividir los datos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de regresión lineal
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = linreg.predict(X_test)

# Evaluar el modelo utilizando el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular el Error Absoluto Medio (MAE) y el Coeficiente de Determinación (R^2)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R^2 Score:", r2)


```

```python
'''
Mean Squared Error (MSE): El MSE mide la diferencia promedio entre las predicciones del modelo y los valores reales. 
En este caso, es aproximadamente 0.0001596. Cuanto más cercano a cero, mejor es el ajuste del modelo.

Mean Absolute Error (MAE): El MAE evalúa el error absoluto promedio entre las predicciones y los valores reales. 
En este caso, es aproximadamente 0.01057. Al igual que con el MSE, cuanto más bajo sea el MAE, mejor se ajustará 
el modelo a los datos.

R^2 Score (coeficiente de determinación): El R^2 indica la proporción de la variabilidad en los datos que se explica por el modelo. 
En este caso, el R^2 es aproximadamente 0.5407, lo que significa que el modelo explica aproximadamente el 54.07% de la variabilidad 
en los datos. El R^2 varía entre 0 y 1, donde un valor más cercano a 1 indica un mejor ajuste del modelo a los datos.
'''
```

```python
# Algoritmo de aprendizaje supervisado (k-Vecinos más Cercanos (k-NN))

'''
El algoritmo k-Vecinos más Cercanos (k-NN) es un método simple y versátil para clasificación y regresión 
en aprendizaje supervisado. La idea principal del k-NN es predecir etiquetas de clase (en clasificación) 
o valores numéricos (en regresión) para nuevas instancias basándose en los ejemplos más cercanos en el 
conjunto de entrenamiento.
'''

import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_vals = [6.290303913, 6.278329421, 6.26393719, 6.250503559, 6.247197047, 6.23061541, 6.212712849, 6.315319592, 6.330537796, 6.337816543, 6.252978384, 6.253342459, 6.258716136, 6.256649006]
Y_vals = [-75.56469784, -75.56942353, -75.56334835, -75.56822247, -75.56970975, -75.57554709, -75.57801847, -75.55528241, -75.55338124, -75.54431178, -75.582943, -75.58827777, -75.59773582, -75.61341926]

X = np.array(X_vals).reshape(-1, 1)  # Convertir X_vals en un array 2D
y = np.array(Y_vals)  # Convertir Y_vals en un array de NumPy

# Dividir los datos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de k-NN (k=3)
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = knn.predict(X_test)

# Evaluar el modelo utilizando el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calcular el Error Absoluto Medio (MAE) y el Coeficiente de Determinación (R^2)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R^2 Score:", r2)
```

```python
'''
Mean Squared Error (MSE): 7.359e-05. Un valor de MSE más cercano a cero indica un mejor ajuste del modelo. En este caso, 
el MSE es pequeño, lo que sugiere que el modelo predice los valores objetivo con un error razonablemente bajo.

Mean Absolute Error (MAE): 0.00638. Similar al MSE, un MAE más bajo indica un mejor ajuste del modelo. Este valor implica 
que el error absoluto promedio entre las predicciones y los valores reales es de aproximadamente 0.00638.

R^2 Score: 0.7882. Este coeficiente de determinación varía entre 0 y 1, donde un valor más cercano a 1 indica un mejor 
ajuste del modelo a los datos. Un R^2 de 0.7882 sugiere que el modelo explica aproximadamente el 78.82% de la variabilidad 
en los datos, lo que indica un buen ajuste del modelo.

En general, estos resultados muestran que el modelo proporciona un ajuste razonablemente bueno a los datos. Puede comparar 
estos resultados con los de otros modelos para elegir el que mejor se ajuste a sus necesidades. Dado que el R^2 es bastante 
alto (78.82%), podríamos considerar que este modelo está funcionando bien en este conjunto de datos. Sin embargo, siempre 
es útil probar diferentes algoritmos y realizar ajustes de hiperparámetros para optimizar aún más el rendimiento del modelo.
'''
```
